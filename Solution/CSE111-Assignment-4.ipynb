```python
#Task 2
class Customer:
    def __init__(self,name):
        self.name=name
    def greet(self,nam=None):
        if nam!=None:
            print("Hello",nam+"!")
        else:
            print("Hello!")
    def purchase(self,*items):
        print(self.name,"you purchased",len(items),"item(s)")
        for i in items:
            print(i)
customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
        
```

```python
#Task 3
class Panda:
    def __init__(self,name,gender,age):
        self.name=name
        self.gender=gender
        self.age=age
    def sleep(self,duration=None):
        self.duration=duration
        if self.duration==None:
            return self.name+"'s' duration is unknown and thus should have only bamboo leaves"
        elif self.duration>=3 and self.duration<=5:
            return self.name+" sleeps "+ str(self.duration)+ " hours daily and should have Mixed Vegetables"
        elif self.duration>=6 and self.duration<=8:
            return self.name+" sleeps "+ str(self.duration)+ " hours daily and should have Eggplant & Tofu"
        elif self.duration>=9 and self.duration<=11:
            return self.name+" sleeps "+ str(self.duration)+ " hours daily and should have Broccoli Chicken"
panda1 = Panda("Kunfu","Male", 5)
panda2=Panda("Pan Pan","Female",3)
panda3=Panda("Ming Ming","Female",8)
print("{} is a {} Panda Bear who is {} yearsold".format(panda1.name,panda1.gender,panda1.age))
print("{} is a {} Panda Bear who is {} yearsold".format(panda2.name,panda2.gender,panda2.age))
print("{} is a {} Panda Bear who is {} yearsold".format(panda3.name,panda3.gender,panda3.age))
print("===========================")
print(panda2.sleep(10))
print(panda1.sleep(4))
print(panda3.sleep())
```

```python
class ParcelKoro:
    def __init__(self,name="No name set",weight=0):
        self.name=name
        self.product_weight=weight
    def calculateFee(self,location=None):
        #self.loca=location
        self.total=(self.product_weight*20)
        if location==None:
            self.loca_charge=50
        else:
            self.loca_charge=100
        if self.total!=0:
            self.total+=self.loca_charge
    def printDetails(self):
        print("Name:",self.name)
        print("Weight:",self.product_weight)
        print("Total fee:",self.total)
print("**********************")
p1 = ParcelKoro()
p1.calculateFee()
p1.printDetails()
print("**********************")
p2 = ParcelKoro('Bob The Builder')
p2.calculateFee()
p2.printDetails()
print("----------------------------")
p2.product_weight = 15
p2.calculateFee()
p2.printDetails()
print("**********************")
p3 = ParcelKoro('Dora The Explorer', 10)
p3.calculateFee('Dhanmondi')
p3.printDetails()
```

```python
#Task 13
class Batsman:
    def __init__(self,name,runscr,ballsfaced=None):
        if ballsfaced==None:
            self.name= "New Batsman"
            self.runscr=name
            self.ballsfaced=runscr
        else:
            self.name=name
            self.runscr=runscr
            self.ballsfaced=ballsfaced
    def setName(self,newname):
        self.name=newname
    def battingStrikeRate(self,battingSR=0):
        self.battingSR=int(self.runscr)/int(self.ballsfaced)*100
        return self.battingSR
    def printCareerStatistics(self):
        print("Name:",self.name)
        print("Runs Scored:",self.runscr,"Balls Faced",self.ballsfaced)
b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```

```python
#Task 15
class Account:
    def __init__(self,name="Default Account",balance=0):
        self.name=name
        self.balance=balance
    def details(self):
        print(self.name)
        return float(self.balance)
    def withdraw(self,balance):
        self.balance -= float(balance)
        if self.balance>3070:
            print("Withdraw successfull! New balance is:", self.balance)
        else:
            self.balance+=float(balance)
            print("Sorry, Withdraw unsuccessful! The account balance after deducting withdraw amount is equal to or less than minimum.")
a1=Account()
print(a1.details())
print("------------------------")
a1.name = "Oliver"
a1.balance = 10000.0
print(a1.details())
print("------------------------")
a2 = Account("Liam")
print(a2.details())
print("------------------------")
a3 = Account("Noah",400)
print(a3.details())
print("------------------------")
a1.withdraw(6930);
print("------------------------")
a2.withdraw(600);
print("------------------------")
a1.withdraw(6929)
```
